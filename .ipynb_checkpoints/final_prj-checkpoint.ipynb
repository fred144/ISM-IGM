Here we'll look at two distinct IMFs



````python
masses = np.geomspace(0.1, 100, 100)


def salp(m, gamma=2.3):
    return m**-gamma


def kroupa(m):
    output = np.zeros_like(m)
    mask1 = (m >= 0.1) & (m < 0.5)
    mask2 = m >= 0.5
    output[mask1] = 2 * m[mask1] ** -1.3
    output[mask2] = m[mask2] ** -2.3
    return output


def top_heavy1(m, alpha=1.7):
    output = np.zeros_like(m)
    mask1 = (m >= 0.1) & (m < 0.5)
    mask2 = m >= 0.5
    output[mask1] = 2 * m[mask1] ** -1.3
    output[mask2] = alpha * m[mask2] ** -alpha
    return output


def top_heavy2(m, alpha):
    output = np.zeros_like(m)
    mask1 = (m >= 0.1) & (m < 0.5)
    mask2 = m >= 0.5
    output[mask1] = 2 * m[mask1] ** -1.35
    output[mask2] = 1.2 * m[mask2] ** -alpha
    return output


fig, ax = plt.subplots(1, 1, dpi=120, figsize=(4, 4))
ax.plot(masses, salp(masses), lw=3, color="rebeccapurple", label="Salpeter 1955")
ax.plot(masses, kroupa(masses), lw=3, color="crimson", label="Kroupa 2001")
ax.plot(
    masses,
    top_heavy2(masses, alpha=2.1),
    lw=3,
    color="grey",
    ls=":",
    label=r"Top Heavy, $\alpha_2 = 2.1$",
)
ax.plot(
    masses,
    top_heavy1(masses, alpha=1.7),
    lw=3,
    color="black",
    ls="--",
    label=r"Top Heavier, $\alpha_2 = 1.7$",
)
ax.set(
    xscale="log",
    yscale="log",
    xlim=(0.1, 100),
    ylim=(1e-3, 200),
    xlabel=r"Stellar Mass [M$_\odot$]",
    ylabel=r"$\xi_0 N$",
)
ax.legend()

plt.savefig("imf.png")
plt.show()
````

we can wrap this in a function so that we only return the specturm at approximately wanted_time

while we're at it we can also calculate being able to see how it changes if we place it to a given redshift (z) for a given observed Flux $F_{\rm obs}$ and wavelength $\lambda$

$$
\lambda_{\rm obs} = \lambda(1+z) 
$$

$$
F_{\rm obs} = \dfrac{L}{ 4 \pi D_L^2}
$$

for a given restframe luminosity (which we have from starburst99), luminosity distance, which depends on redshift.

````python
def get_spectrum(file, wanted_tmyr, type="total"):
    if type == "total":
        time, wvlngth, lum = np.loadtxt(
            file, unpack=True, usecols=(0, 1, 2), skiprows=6
        )
    elif type == "stellar":
        time, wvlngth, lum = np.loadtxt(
            file, unpack=True, usecols=(0, 1, 3), skiprows=6
        )
    elif type == "nebular":
        time, wvlngth, lum = np.loadtxt(
            file, unpack=True, usecols=(0, 1, 4), skiprows=6
        )
    else:
        print("type on valid")
        exit()

    # convert time
    tmyr = time / 1e6
    wav_ang = wvlngth * u.Angstrom

    # since it is a table, get the closest Myr to it
    wanted_idx = np.argmin(np.abs(np.unique(tmyr) - wanted_tmyr))
    closest_time = np.unique(tmyr)[wanted_idx]
    # print("Closest time to wanted time: {} Myr".format(closest_time))
    valid_time_mask = tmyr == closest_time

    wav = wav_ang[valid_time_mask]
    log_total_luminosity = lum[valid_time_mask]
    luminosity = 10**log_total_luminosity * (u.erg / u.s / u.Angstrom)

    return wav.to("micron").value, luminosity.value
````

````python
tmyr = 1
redshift = 7


def observed_flux_calc(wav, lum, redshift):
    cosmology = FlatLambdaCDM(H0=70 * u.km / u.s / u.Mpc, Tcmb0=2.725 * u.K, Om0=0.3)
    lum_distance = cosmology.luminosity_distance(redshift).to("cm")
    observed_flux = lum / (4 * np.pi * lum_distance**2)
    observed_wavlength = wav * (1 + redshift)
    return observed_wavlength, observed_flux


wav, lum_total = get_spectrum(
    "kroupa_metal1e-4/kroupa_metal1e-4.spectrum1", wanted_tmyr=tmyr, type="total"
)
observed_wavlength, observed_flux = observed_flux_calc(
    wav, lum_total, redshift=redshift
)

_, lum_stellar = get_spectrum(
    "kroupa_metal1e-4/kroupa_metal1e-4.spectrum1", wanted_tmyr=tmyr, type="stellar"
)
_, observed_flux_stellar = observed_flux_calc(wav, lum_stellar, redshift=redshift)

_, lum_nebular = get_spectrum(
    "kroupa_metal1e-4/kroupa_metal1e-4.spectrum1", wanted_tmyr=tmyr, type="nebular"
)
_, observed_flux_nebular = observed_flux_calc(wav, lum_nebular, redshift=redshift)

fig, ax = plt.subplots(
    2,
    1,
    dpi=200,
    figsize=(8, 4),
)
plt.subplots_adjust(hspace=0.4)
# plot from micron to Angstrom
ax[0].plot(wav * 1e4, lum_total, color="crimson", label="total")
ax[0].plot(wav * 1e4, lum_stellar, color="rebeccapurple", label="stellar", ls="--")
ax[0].plot(
    wav * 1e4,
    lum_nebular,
    color="salmon",
    label="nebular",
)
ax[0].set(
    yscale="log",
    ylim=(1e35, 5e40),
    xlim=(100, 8e3),
    xlabel=r"$\lambda_{\rm rest \, frame}$ [$\mathrm{\AA}$]",
    ylabel=r"$L_{\lambda {\rm rest \, frame}}$ [${\rm erg \: s^{-1} \: \AA^{-1}}$]",
)


ax[1].plot(observed_wavlength, observed_flux, color="crimson", label="total")
ax[1].plot(
    observed_wavlength,
    observed_flux_stellar,
    color="rebeccapurple",
    label="stellar",
    ls="--",
)
ax[1].plot(
    observed_wavlength,
    observed_flux_nebular,
    color="salmon",
    label="nebular",
)
ax[1].set(
    xlim=(0.1, 6),
    yscale="log",
    ylim=(1e-25, 2e-20),
    xlabel=r"$\lambda_{\rm obs}$ [$\mu$m]",
    ylabel=r"$F_{\lambda {\rm obs}}$ [${\rm erg \: s^{-1} \: cm^{-2} \AA^{-1}}$]",
)

ax[0].text(
    0.05,
    0.95,
    "age = {} Myr".format(tmyr),
    va="top",
    ha="left",
    transform=ax[0].transAxes,
)
ax[1].text(
    0.95,
    0.95,
    "placed at z = {}".format(redshift),
    va="top",
    ha="right",
    transform=ax[1].transAxes,
)

ax[0].legend(title="Kroupa")
plt.savefig("luminosity_correction.png", bbox_inches="tight")
plt.show()
````

````python
def m_ab(wav, lum, redshift):
    """
    conversion taken form garcia+23
    """

    cosmology = FlatLambdaCDM(H0=70 * u.km / u.s / u.Mpc, Tcmb0=2.725 * u.K, Om0=0.3)
    lum_distance = cosmology.luminosity_distance(redshift).to("Gpc").value
    M_ab = 15.65 - 2.5 * np.log10(lum / 1e39)
    m_ab = M_ab + 5 * np.log10(lum_distance / 100) + 50
    observed_wavlength = wav * (1 + redshift)
    return observed_wavlength, m_ab


tmyr = 0.01
wav_k, lum_total_k = get_spectrum(
    "kroupa_metal1e-4/starburst99.stb99", wanted_tmyr=tmyr, type="total"
)
obs_k, mab_k = m_ab(wav_k, lum_total_k, 7)

wav_t, lum_total_t = get_spectrum(
    "topheavy_metal1e-4/starburst99.stb99", wanted_tmyr=tmyr, type="total"
)
obs_t, mab_t = m_ab(wav_t, lum_total_t, 7)

wav_tr, lum_total_tr = get_spectrum(
    "topheavier_metal1e-4/starburst99.stb99", wanted_tmyr=tmyr, type="total"
)
obs_tr, mab_tr = m_ab(wav_tr, lum_total_tr, 7)

fig, ax = plt.subplots(2, 1, dpi=200, figsize=(8, 2), sharex=True, sharey=True)
plt.subplots_adjust(hspace=0, wspace=0)
ax[0].plot(obs_k, mab_k, label="kroupa")
ax[0].plot(obs_t, mab_t, label="top heavy")
ax[0].plot(obs_tr, mab_tr, label="top heavier")

tmyr = 5
wav_k, lum_total_k = get_spectrum(
    "kroupa_metal1e-4/starburst99.stb99", wanted_tmyr=tmyr, type="total"
)
obs_k, mab_k = m_ab(wav_k, lum_total_k, 7)

wav_t, lum_total_t = get_spectrum(
    "topheavy_metal1e-4/starburst99.stb99", wanted_tmyr=tmyr, type="total"
)
obs_t, mab_t = m_ab(wav_t, lum_total_t, 7)

wav_tr, lum_total_tr = get_spectrum(
    "topheavier_metal1e-4/starburst99.stb99", wanted_tmyr=tmyr, type="total"
)
obs_tr, mab_tr = m_ab(wav_tr, lum_total_tr, 7)

ax[1].plot(obs_k, mab_k, label="kroupa")
ax[1].plot(obs_t, mab_t, label="top heavy")
ax[1].plot(obs_tr, mab_tr, label="top heavier")

ax[0].set(xlim=(0.1, 3), ylim=(61, 70))
ax[0].set(ylabel=r"$m_{\rm AB}$")
ax[1].set(xlabel=r"$\lambda_{\rm obs}$ [$\mu$m]", ylabel=r"$m_{\rm AB}$")

ax[0].invert_yaxis()
ax[0].legend()

ax[0].text(
    0.05,
    0.05,
    "age = {} Myr".format(0.01),
    va="bottom",
    ha="left",
    transform=ax[0].transAxes,
)
ax[1].text(
    0.05,
    0.05,
    "age = {} Myr".format(5),
    va="bottom",
    ha="left",
    transform=ax[1].transAxes,
)


plt.savefig("spectrum_mag.png", bbox_inches="tight")
plt.show()
````

# Running Cloudy

````python
import pyCloudy as pc
import os

cosmology = FlatLambdaCDM(H0=70 * u.km / u.s / u.Mpc, Tcmb0=2.725 * u.K, Om0=0.3)
````

````python
def run_cloudy(
    dir,
    model_name,
    dens,
    metallicity,
    stb99_ascii,
    age,
    set_distance: tuple,
    imf,
    Teff=36870.0,
):
    """
    run cloudy using Starburst99 tables, based on Pset1 Problem3
    dir = path to where outputs are to be stored
    model_name = name of the model to be place (as the prefix) of the files in dir
    dens = log10(dens) in cm^-3, uniform
    metals =  hii region abundance in Zsun
    stb99_ascii = path to the starburst99.ascii
    age = age in years passed as a str; e.g. "1e6" for 1 Myr
    set_distance = unit can be 'kpc', 'Mpc', 'parsecs', 'cm'; e.g. (1, "Mpc")
    """
    pc.config.cloudy_exe = "/home/fabg/Cloudy/c23.01/source/cloudy.exe"

    if not os.path.exists(dir):
        print("> creating container", dir)
        os.makedirs(dir)
    full_model_name = os.path.join(dir, model_name)

    abundance_predef = "hii"  # Use pre-defined hii region abundances
    r_min = 5e16  # cm Inner radius (use as default)
    # dens = 3. #log cm-3, we'll assume uniform density
    # metals = 1   # Metal scaling for abundances (
    # r_max = 2e18 # cm  # Outer radius (if needed)
    # dist = 1.0  # kpc, only used for final output spectra
    if imf == "kroupa":
        wav, lum_total = get_spectrum(
            "kroupa_metal1e-4/starburst99.stb99",
            wanted_tmyr=float(age) / 1e6,
            type="total",
        )
    elif imf == "topheavy":
        wav, lum_total = get_spectrum(
            "topheavy_metal1e-4/starburst99.stb99",
            wanted_tmyr=float(age) / 1e6,
            type="total",
        )
    elif imf == "topheavier":
        wav, lum_total = get_spectrum(
            "topheavier_metal1e-4/starburst99.stb99",
            wanted_tmyr=float(age) / 1e6,
            type="total",
        )
    else:
        print("Does not exist")
        exit()

    lum = np.trapz(lum_total, wav * 1e4)  # integrated luminosity to feed into cloudy
    print("log luminosity", np.log10(lum))
    # fmt: off
    options = (
        "no molecules", "no level2 lines", "no fine opacities",
        "atom h-like levels small", "atom he-like levels small",
        "COSMIC RAY BACKGROUND", "element limit off -8", "print line optical depth",
        "table stars \"{}\" age={} years".format(stb99_ascii, age),
        "luminosity total {}".format(np.log10(lum))
         # "table star \"{}\" 24300 4.3".format(stb99_ascii)
    )
     # fmt: on
    # emission lines that we may be using for analysis of model outputs
    # fmt: off
    emis_tab = [
        "H  1  4861.33A", "H  1  6562.81A", "Ca B  5875.64A", "N  2  6583.45A",
        "N  2  5756.20A", "N  2  6585.30A", "O  1  6300.30A", "O  2  3726.03A",
        "O  2  3727.10A", "O  2  3729.80A", "O  2  3728.81A", "S  2  6716.44A",
        "S  2  6718.30A", "S  2  6730.82A", "S  2  6732.70A", "O  3  4364.40A",  
        "O  3  5006.84A", "O  3  5008.20A", "BLND  4363.00A", "Cl 3  5517.71A",
        "Cl 3  5537.87A", "O  1  63.1679m", "O  1  145.495m", "C  2  157.636m",
    ]
    # fmt: on

    # Defining the object that will manage the input file for Cloudy
    c_input = pc.CloudyInput(full_model_name)

    # Filling the object with the parameters
    # Defining the ionizing SED: Effective temperature and luminosity.
    # The lumi_unit is one of the Cloudy options,
    # like "luminosity solar", "q(H)", "ionization parameter", etc...

    qH = 48.75  # Rate of production of Hydrogen ionizing photons in log photons s^-1
    # c_input.set_BB(Teff=Teff, lumi_unit="q(H)", lumi_value=qH)

    # elif config == "seyfert":
    #     c_input.set_star()
    #     c_input.set_star(
    #         SED="table star",
    #         SED_params="",
    #         lumi_unit="ionization parameter",
    #         lumi_value=-2,
    #     )

    # Defining the density and profile
    c_input.set_cste_density(dens)  # constant density

    # Defining the inner radius.
    # A second parameter could be the outer radius (matter-bounded nebula).
    # c_input.set_radius(r_in=np.log10(r_min),r_out=np.log10(r_max))
    c_input.set_radius(r_in=np.log10(r_min))

    # Use pre-defined hii region abundances
    c_input.set_abund(predef=abundance_predef, nograins=True, metals=metallicity)

    # Other options from above
    c_input.set_other(options)

    # (0) for no iteration, () for one iteration, (N) for N iterations.
    c_input.set_iterate(10)

    # () or (True) : sphere, or (False): open geometry.
    c_input.set_sphere(True)

    c_input.set_emis_tab(emis_tab)

    c_input.set_distance(*set_distance, linear=True)
    # Writing the Cloudy inputs.
    # to_file for writing to a file (named by full_model_name). verbose to print on the screen.
    c_input.print_input(to_file=True, verbose=False)

    # Printing some message to the screen
    pc.log_.message("Running {0}".format(model_name), calling=model_name)

    # Running Cloudy with a timer. Here we reset it to 0.
    pc.log_.timer("Starting Cloudy", quiet=True, calling=model_name)
    c_input.run_cloudy()
    pc.log_.timer("Cloudy ended after seconds:", calling=model_name)
````
